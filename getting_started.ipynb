#### Select Queries
The most basic command in sql is the `select` command. It is used whenever we want to query the database.\
Let's have a look at how to write a basic select query
```sql
select * from sample
```
- the keyword `select` means this is a query
- Afterward, the star symbol `*` means we want to select *all* of the columns from the table.
- the keyword `from` followed by the name of the table we want to query (in this case, the table `sample`)

note: SQL is not case sensitive, so both `SELECT` and `select` are valid. It used to be considered best practice to use upper case for KEYWORDS and lower case for everything else, but I will avoid this convention as I find it hurts readability

try running a select query:

````mysql
select * from sample where calculated_host_listings_count
````

Let's look at what we got:
- there are 10 records (or rows) of airbnb apartments in New York City
- each record have 16 different columns of values - name of the place, name of the host, price, et cetera
